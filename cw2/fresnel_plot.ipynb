```python
from math import sin, cos, asin, atan
from functools import partial as pt
import matplotlib.pyplot as plt
import numpy as np
import operator

eta_air = 1.00293
eta_medium = 1.5

theta = np.linspace(0,math.pi/2,1000)

asin_param = np.linspace(0,1,1000)
asin_param = map(lambda x: x / eta_medium * eta_air, asin_param)

theta2 = np.linspace(0, asin(0.6666),1000)
```

```python
def snell_thetaT(thetaI, etaI, etaT):
    return asin(sin(thetaI)*etaI/etaT)

def Rs(thetaI, etaI, etaT):
    thetaT = snell_thetaT(thetaI, etaI, etaT)
    numer = etaI * cos(thetaI) - etaT * cos(thetaT)
    denom = etaI * cos(thetaI) + etaT * cos(thetaT)
    return abs(numer / denom)**2

def Rp(thetaI, etaI, etaT):
    thetaT = snell_thetaT(thetaI, etaI, etaT)
    numer = etaI * cos(thetaT) - etaT * cos(thetaI)
    denom = etaI * cos(thetaT) + etaT * cos(thetaI)
    return abs(numer / denom)**2

Rs_partial = pt(Rs, etaI = eta_air, etaT = eta_medium)
Rp_partial = pt(Rp, etaI = eta_air, etaT = eta_medium)
coeff_s = map(Rs_partial, theta)
coeff_p = map(Rp_partial, theta)

minRp_index, minRp_value = min(enumerate(coeff_p), key=operator.itemgetter(1))

print 'Actual Brewster Angle: ' + str(theta[minRp_index])
print 'Theretical Brewster Angle: ' + str(atan(eta_medium/eta_air))

fig = plt.figure()
ax = fig.add_subplot(111)
ax.grid(False)
ax.plot(theta, coeff_s,'-', label='R_s')
ax.plot(theta, coeff_p,'-', label='R_p')
ax.set_xlabel('Angle of Incidence, theta_i')
ax.set_ylabel('Reflection Coefficient')
ax.set_title('Reflection with eta_t = 1.5')
ax.vlines(x=theta[minRp_index], ymin=0.0, ymax=1.0, colors='k', linestyles='dashed', label='Brewster Angle')
##ax.set_xlim([0,math.pi/2])
legend()
plt.show()
```

```python
Rs_new = pt(Rs, etaI = eta_medium, etaT = eta_air)
Rp_new = pt(Rp, etaI = eta_medium, etaT = eta_air)

coeff_new_s = map(Rs_new, theta2)
coeff_new_p = map(Rp_new, theta2)
max_new_Rp, max_new_Rp_value = max(enumerate(coeff_new_p), key=operator.itemgetter(1))
minRp_index, minRp_value = min(enumerate(coeff_p), key=operator.itemgetter(1))

fig2 = plt.figure()
ax2 = fig2.add_subplot(111)
ax2.grid(False)
ax2.plot(theta2, coeff_new_s, '-', label='R_s')
ax2.plot(theta2, coeff_new_p, '-', label='R_p')
ax2.set_xlabel('Angle of Incidence, theta_i')
ax2.set_ylabel('Reflection Coefficient')
ax2.set_title('Reflection from eta_t = 1.5 to eta_air = ' + str(eta_air))
ax2.vlines(x=theta2[-1], ymin=0.0, ymax=1, colors='k', linestyles='dashed', label='Critical Angle')

ax2.legend(loc=2)
plt.show()
```
